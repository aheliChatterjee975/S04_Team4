```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("spotify_tracks.csv")

# List of numeric features
numeric_cols = ['popularity', 'duration_ms', 'danceability', 'energy', 'loudness',
                'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']

# Histograms
sns.set(style="whitegrid")
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(18, 15))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)

# Remove any extra subplot
if len(axes) > len(numeric_cols):
    fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 10))
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
def remove_outliers_iqr(df, features):
    for col in features:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower) & (df[col] <= upper)]
    return df

df_cleaned = remove_outliers_iqr(df[numeric_cols].copy(), numeric_cols)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
import numpy as np

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cleaned)

# Determine optimal k (Elbow Method)
sse = []
k_range = range(2, 11)
for k in k_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    sse.append(km.inertia_)

plt.plot(k_range, sse, marker='o')
plt.title('Elbow Method - Optimal K')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()

# Silhouette Analysis
silhouette_scores = []
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    silhouette_scores.append(silhouette_score(X_scaled, labels))

plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Scores')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()

# Try different models
kmeans_labels = KMeans(n_clusters=5, random_state=42).fit_predict(X_scaled)
dbscan_labels = DBSCAN(eps=1.5, min_samples=10).fit_predict(X_scaled)
agg_labels = AgglomerativeClustering(n_clusters=5).fit_predict(X_scaled)

# Compare silhouette scores
print("KMeans Silhouette Score:", silhouette_score(X_scaled, kmeans_labels))
print("DBSCAN Silhouette Score:", silhouette_score(X_scaled, dbscan_labels))
print("Agglomerative Silhouette Score:", silhouette_score(X_scaled, agg_labels))
```

```python
df_final = df.loc[df_cleaned.index].copy()
df_final['cluster'] = kmeans_labels

# Function to get recommendations
def recommend_songs(song_names, n_recommendations=5):
    recommended = []
    for song in song_names:
        match = df_final[df_final['track_name'].str.lower() == song.lower()]
        if match.empty:
            print(f"Song '{song}' not found.")
            continue
        cluster = match['cluster'].values[0]
        cluster_songs = df_final[(df_final['cluster'] == cluster) &
                                 (~df_final['track_name'].str.lower().isin([song.lower()]))]
        recs = cluster_songs.sample(n=min(n_recommendations, len(cluster_songs)))
        recommended.extend(recs[['track_name', 'artists', 'track_genre']].values.tolist())
    return recommended


favorite_songs = ["Hold On", "To Begin Again"]
recommendations = recommend_songs(favorite_songs, n_recommendations=3)

for i, rec in enumerate(recommendations, 1):
    print(f"{i}. {rec[0]} by {rec[1]} ({rec[2]})")
```

```python
# Interactive feedback collection
def get_feedback(recommendations):
    feedback = {}
    for i, (track, artist, genre) in enumerate(recommendations, 1):
        print(f"{i}. {track} by {artist} ({genre})")
        rating = input("Did you like this recommendation? (yes/no): ").strip().lower()
        feedback[(track, artist)] = rating
    return feedback

# Collecting feedback after showing recommendations
feedback_results = get_feedback(recommendations)
print("\nUser Feedback Summary:")
for song, feedback in feedback_results.items():
    print(f"{song[0]} by {song[1]}: {feedback}")
```

```python
# 4. INNOVATION COMPONENT: Sentiment Analysis
# Simulate some reviews for recommended songs (in exteneded cases, this would be integrated into APIs or user input)
from textblob import TextBlob

simulated_reviews = {
    "Hold On": "I love the emotional depth and the melody is just beautiful.",
    "To Begin Again": "It felt a bit repetitive, not my favorite.",
    "New Light": "Super catchy and upbeat. Loved it!",
    "Lose Control": "A bit too loud for my taste.",
    "Afterglow": "Beautiful lyrics and soothing rhythm."
}

# Function to analyze sentiment
def analyze_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity  # -1 (negative) to 1 (positive)

# Apply sentiment to the simulated reviews
print("\nSentiment Analysis of Sample Reviews:")
for song, review in simulated_reviews.items():
    score = analyze_sentiment(review)
    sentiment = "Positive" if score > 0 else "Negative" if score < 0 else "Neutral"
    print(f"{song}: {sentiment} (Score: {score:.2f})")
```
